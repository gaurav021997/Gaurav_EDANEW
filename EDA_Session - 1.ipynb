# Creat Dataframes Using List

```python
name = ['Ramesh','Suresh','sathish']
age = [30,35,40]
name,age
```

**Step-1**

$Create$ $data$ $frame$

```python
pd.DataFrame()
```

**Step-2**

$provide$ $data$

```python
pd.DataFrame(zip(name,age))
```

**Step-3**

$provide$ $column$

```python
# provied columns
data=zip(name,age)
cols=['Name','Age']
pd.DataFrame(data,columns=cols)
#pd.DataFrame(zip(name,age),columns=['Name','Age'])
```

**Step-4**

$provide$ $index$

```python
data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
pd.DataFrame(data,
            columns=cols,
            index=ind)
```

**Step-5**

$add$ $new$ $column$

```python
name = ['Ramesh','Suresh','sathish']
age = [30,35,40]

data=zip(name,age)
cols=['Name','Age']
ind=['A','B','C']
df=pd.DataFrame(data,columns=cols,index=ind)
df
```

- if you want to add a new column

- df['new column']

- you need to have a list which is having some elements

- that elements need to equal to number of rows

- city_names=['Hyd','blr','Chennai']

- df['city']=city_names

```python
city_names=['Hyd','Blr','Chennai']
df['city']=city_names
df
```

**Step-6**

$update$ $the$ $Exsiting$ $Column$

 - if you want to create new column or update the old column
 - both are same way
    

```python
df['Name']=['Swamy','Asif','Sathwik']
df
```

**Step-7**

$drop$ $the$ $column$

- In order to drop the column

- We need to use drop method

- It takes 3 parameters
    
    - drop column or now
    
    - mention the column name
    
    - axis
    
        - axis=1 referance as column
        
        - axis=0 referance as row
        
    - You wantto creat a new DataFrame or
    
    - You want overwrite the exciting dataframe
    
        - inplace = True
    


```python
df.drop('city',        # Column Name
        axis=1,        # Column
       inplace=True)   # overwrite the same
df
```

```python
name=['Ramesh','Suresh','Sathish']
age=[30,35,40]

df=pd.DataFrame(zip(name,age),
                columns=['Name','Age'],
                index=['A','B','C'])

city_names=['Hyd','Blr','Chennai']
df['city']=city_names

df.drop('city',
        axis=1,
       inplace=True)
df
                
```

$step-8$

Drop rows

```python
df.drop('C', # column name
        axis=0, # Column
        inplace=True) # overwrite the same
df
```

$step-9$

$save$ $the$ $data$ $frame$

```python
df.to_csv("output.csv")
# while saving index consider as extra column
df.to_excel("output.xlsx")
```

```python
#read output csv
pd.read_csv("output.csv")
```
